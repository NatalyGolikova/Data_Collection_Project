```python
import requests
import json
import csv

URL = 'https://api.kinopoisk.dev/v1.4/movie'
TOKEN = 'DTFGKW2-7KB4JFR-JZND1GT-ZGV52Q7'

session = requests.Session()
session.headers.update({'accept': 'application/json', 'X-API-KEY': TOKEN})

result = []
page=1
while page<=4:
  response = session.get(URL, params={
      'selectFields': ['name', 'movieLength', 'countries', 'rating'],
      'notNullFields': ['name'],
      'type': ['movie'],
      'year': 2000,
      'genres.name': ['комедия'],
      'sortField': 'rating.kp',
      'sortType': [-1],
      'limit': 250,
      'page': page,
  })

  if response:
    docs = response.json().get('docs', [])
    for item in docs:
      item['countries'] = [country['name'] for country in item['countries']]
      item['rating'] = item['rating']['kp']
    result += docs

  page+=1

with open('Movies_Kinopoisk_API.json', 'w', encoding='utf-8') as f:
  json.dump(result, f)
```
