```python
import requests
import time
import csv

URL_AREAS = 'https://api.hh.ru/areas'
URL_EMPLOYERS = 'https://api.hh.ru/employers'

response = requests.get(URL_AREAS)
if response.status_code:
  response_data = response.json()
  obj_ru = next(data for data in response_data if data['name']=='Россия')
  altai_id = [area['id'] for area in obj_ru['areas'] if area['name']=='Алтайский край']

print(altai_id)
employers=[]
page=0
while True:
  response_em = requests.get(URL_EMPLOYERS, params={
          'page': page,
          'per_page': 100,
          'area': altai_id,
          'sort_by': 'by_vacancies_open',
          'only_with_vacancies': True
          })
  if response_em.status_code:
    items = response_em.json().get('items', [])
    if len(items)==0: break
    for item in items:
      em_dict = {'id': item['id'],
              'name': item['name'],
              'vacancies_url': item['vacancies_url'],
              'open_vacancies': item['open_vacancies']}
      employers.append(em_dict)
  else:
    print(response_em.status_code, response_em.json())
    break
  page +=1

  time.sleep(0.3)


with open('Employers_Altai.csv', 'w', encoding='utf-8') as f:
  dict_writer = csv.DictWriter(f, employers[0].keys())
  dict_writer.writeheader()
  dict_writer.writerows(employers[:1000])
```
